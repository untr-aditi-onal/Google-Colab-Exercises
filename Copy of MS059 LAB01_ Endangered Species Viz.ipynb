````
# Import all of the libraries you need
from PIL import Image, ImageDraw, ImageFont # Open and manipulate images on the server
from math import sqrt                       # Function to get the square root of a number
import requests                             # Download images
import matplotlib.pyplot as plt             # Make plots with Python

# use a truetype font
fnt = ImageFont.truetype('slkscr.ttf', 21)
bfnt = ImageFont.truetype('slkscr.ttf', 72)
````

````
pandaUrl   = 'https://www.evernote.com/l/ADOw2tfbp7lE8p-f7yxcK74ghHwVuvR4v0IB/image.png'
jaguarUrl  = 'https://www.evernote.com/l/ADNsAD0jWDtGvYq4I85KOUQXz9HDwCN4M-YB/image.png'
gorillaUrl = 'https://www.evernote.com/l/ADOdBwKmLZ5G6q8ClR2-Ru50tWj_v6cgofkB/image.png'
rhinoUrl   = 'https://www.evernote.com/l/ADNAollR6kFPzrtD0ETXuZ9ycRSraoijuCYB/image.png'
leopardUrl = 'https://www.evernote.com/l/ADMNhBAFfAdL9r0WZRstcqQ9XqIcDuN82b0B/image.png'
orangUrl   = 'https://www.evernote.com/l/ADOG4tgj-WpHVaH-F5O4z4lMJzdSQDzHl10B/image.png'

panda = Image.open(requests.get(pandaUrl, stream=True).raw)
jaguar = Image.open(requests.get(jaguarUrl, stream=True).raw)
gorilla = Image.open(requests.get(gorillaUrl, stream=True).raw)
rhino = Image.open(requests.get(rhinoUrl, stream=True).raw)
orang = Image.open(requests.get(orangUrl, stream=True).raw)
````

# How do you reduce the number of pixels? 

How many giant pandas are left? 

# Set up your count and message. Get the data right!
# At least count (2014 census) there were 1,864 giant pandas living in the wild.

`cnt = 1864`

````
# How do you reduce the number of pixels? 
# Resize it to 50x50 pixels

cnt = 1864
msg = "Giant Pandas."

rows = cols = int(sqrt(cnt))  # Note that rows equal to cols 
                              # because it's a square root and we want a square Image
                              # Also that both are integers!

print(rows,cols)

coarse = panda.resize((cols,rows))
plt.imshow(coarse)
````

# Resize the Image to 400x400 pixels

````
# Bring it back to 400x400 pixels

# yes, you need to call figure() every time!
fig = plt.figure(figsize=(10,10))

coarse400x400 = coarse.resize((400,400))
plt.imshow(coarse400x400)
````

# Yuck!

## Use resample=Image.NEAREST prevents the Python Image Library (PIL) 

````
# Yuck!
# Use resample=Image.NEAREST prevents the Python Image Library (PIL) 
# from smoothing out your image

# yes, you need to call figure() every time!
fig = plt.figure(figsize=(10,10))

coarse400x400 = coarse.resize((400,400),resample=Image.NEAREST)
plt.imshow(coarse400x400)
````

# Add an annotation

````
# Add an annotation

# yes, you need to call figure() every time!
fig = plt.figure(figsize=(10,10))

# Set up your count and message. Get the data right!
# At least count (2014 census) there were 1,864 giant pandas living in the wild.
cnt = 1864
msg = "Giant Pandas."

# Configure the truetype font ([Silkscreen, by Jason Kottke](https://kottke.org/plus/type/silkscreen/)) that you downloaded
fnt = ImageFont.truetype('slkscr.ttf', 21)
bfnt = ImageFont.truetype('slkscr.ttf', 72)

# Set up a Draw object
draw  = ImageDraw.Draw(coarse400x400)
# Draw two lines of text with the specified fill color and font
draw.text( (10,300), str(cnt), fill='#E8CeB7', font=bfnt)
draw.text( (10,360), msg, fill='#E8CeB7', font=fnt)

# Let's see!
plt.imshow(coarse400x400)
````
