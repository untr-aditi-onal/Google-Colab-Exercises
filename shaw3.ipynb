```
# printing the follwoing statement through calling function of print to do it to 
#the string in apostrophes and parantheses
print ("I will now count my chickens:")

# calculating mathematical expression and labeling them hens and roosters accordingly  
print ("Hens", 25 + 30 / 6)
print ("Roosters", 100 - 25 * 3 % 4)

#printing another statement
print ("Now I will count the eggs:")

#calculating using order of operations 
print (3 + 2 + 1 - 5 + 4 % 2 - 1 / 4 + 6)

# evaluating a boolean to see if true or false after calculating
print ("Is it true that 3 + 2 < 5 - 7?")

# printing to test the statement
print (3 + 2 < 5 - 7)

# more printing and calculating
print ("What is 3 + 2?", 3 + 2)
print ("What is 5 - 7?", 5 - 7)

# confirming after the calculation that it's false that 5 is less than -2
print ("Oh, that's why it's False.")

# more printing, transition into more 
print ("How about some more.")

# evaluating boolean expression and asking computer if after using the operators in the condition whether they're T or F
print ("Is it greater?", 5 > -2)
print ("Is it greater or equal?", 5 >= -2)
print ("Is it less or equal?", 5 <= -2)
```
