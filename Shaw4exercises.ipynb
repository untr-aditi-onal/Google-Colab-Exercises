```
# there are 100 cars
cars = 100
# only 4 seats to a car
space_in_a_car = 4.0
# there are 30 drivers
drivers = 30
# there are 90 passengers
passengers = 90
# to calculate the difference between cars that each driver has and how many leftover cars there is do this
cars_not_driven = cars - drivers
# the rest as logically makes sense are the drivers
cars_driven = drivers
# to see how much all the cars being driven can take it can be multiplied by the number of seats per car
carpool_capacity = cars_driven * space_in_a_car
# to see how many ppl can fit in available driven cars calculate the following
average_passengers_per_car = passengers / cars_driven


print ("There are", cars, "cars available.")
print ("There are only", drivers, "drivers available.")
print ("There will be", cars_not_driven, "empty cars today.")
print ("We can transport", carpool_capacity, "people today.")
print ("We have", passengers, "to carpool today.")
print ("We need to put about", average_passengers_per_car, "in each car.")
```
